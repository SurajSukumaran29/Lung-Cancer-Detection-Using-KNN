#IMPORT THE CSV FILE

```python
#import the csv file
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/Datasets/lung_cancer_examples.csv')
df
```

#CHECK FOR ANY MISSING VALUES

```python
#check for missing value
df.isnull().sum()
```

#CHECK THE DATATYPES OF EACH COLUMN

```python
df.dtypes
```

#DROP THE COLUMNS WHICH ARE NOT NECESSARY FOR TRAINING

```python
#drop data columns
df.drop(['Name','Surname'],axis=1,inplace=True)
df
```

#ASSIGN THE FEATURES TO X

```python
X=df.iloc[:,:-1].values
X
```

#ASSIGN THE TARGETS TO Y

```python
y=df.iloc[:,-1]
y
```

#SCALE THE VALUES OF X BETWEEN 0 to 1

```python
#Scale the X_column values to range between 0-1
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(X)
X_scaled=scaler.transform(X)
X_scaled
```

#SPLIT THE TRAINING AND TEST SET

```python
#Splitting the data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.3,random_state=1)
```

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

```python
X_test
```

```python
y_test.values
```

#BUILD THE MODEL USING KNN

```python
#Building the Model
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=3)
model.fit(X_train,y_train)
```

#PREDICT THE TEST FEATURES WITH THE MODEL

```python
#Predicting with Model
y_pred=model.predict(X_test)
y_pred
```

#CALCULATE THE ACCURACY OF THE MODEL

```python
#Calculate the Accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred)*100)
```
